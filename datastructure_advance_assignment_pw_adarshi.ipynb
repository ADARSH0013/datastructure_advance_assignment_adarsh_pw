```python
#1- Write a code to reverse a strin.

original_string= "adarsh"
reversed_string= original_string[::-1]
print(reversed_string)


#2- Write a code to count the number of vowels in a string

string = "datascience!"
vowels = "aeiouAEIOU"

count = sum(string.count(vowel) for vowel in vowels)
print(count)

#3- Write a code to check if a given string is a palindrome or not


my_str = 'racecaR'
my_str = my_str.casefold()
rev_str = reversed(my_str)

if list(my_str) == list(rev_str):
   print("The string is a palindrome.")
else:
   print("The string is not a palindrome.")


#4- Write a code to check if two given strings are anagrams of each other.

string1= "Race"
string2= "Care"
#convert both the string into lower case 
string1= string1.lower()
string2= string2.lower()
if(len(string1) == len(string2)):
#sorting
   sorted_string1 = sorted(string1)
   sorted_string2 = sorted(string2)

   if(sorted_string1 == sorted_string2):
      print("anagram")
   else:
      print("not anagram")



#5- Write a code to find all occurrences of a given substring within another string.

#using re.findall method

import re 
str = " i am i "
substr = "i"
count= len(re.findall(substr, str))
print("number of substring", count)


#6-  Write a code to perform basic string compression using the counts of repeated characters

def compress_string(s):
    compressed = []
    count = 1

    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            compressed.append(s[i - 1] + str(count))
            count = 1
    
    # Adding the last character's count
    compressed.append(s[-1] + str(count))  
    
    compressed_str = ''.join(compressed)
    
    return compressed_str if len(compressed_str) < len(s) else s

# Test the function
input_string = input("Enter a string to compress: ")
compressed_result = compress_string(input_string)
print("Compressed string:", compressed_result)



#7- Write a code to determine if a string has all unique characters

def check_unique(str):
  for i in range(len(str)):
    for j in range(i + 1,len(str)):
      if(str[i] == str[j]):
        return False
  return True

str = input("enter a string : ")
if(check_unique(str)):
  print("contains unique characters")
else:
  print("contains duplicate characters")


#8-  Write a code to convert a given string to uppercase or lowercase
#converting to uppercase
string1= "i am adarsh" 
print(string.upper())
#converting to lowercase
string= "I AM ADARSH"
print(string.lower())


#9- Write a code to count the number of words in a string

string= "i am learning python from pw"
word_count= string.count(" ")+1
print(word_count)


#10- Write a code to concatenate two strings without using the + operator.

string1 = "i am learning"
string2 = "python"
concatenate_string = " ".join([ string1 , string2])
print(concatenate_string)


#11- Write a code to remove all occurrences of a specific element from a list.

list= [1,2,3,1,4,1,4,1,5,6,7]
while(list.count(1)):
   list.remove(1)
print(list)


#12-  Implement a code to find the second largest number in a given list of integers.

list= [10,11,12,13,14]
list.sort()
print("second largest:", list[-2])


#13- Create a code to count the occurrences of each element in a list and return a dictionary with elements as keys and their counts as values

def count_occurrences(lst):
    count_dict = {}
    for element in lst:
        if element in count_dict:
            count_dict[element] += 1
        else:
            count_dict[element] = 1
    return count_dict
my_list = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
result = count_occurrences(my_list)
print(result)


#14-  Write a code to reverse a list in-place without using any built-in reverse functions.

a= [1,2,3,4,5]
start,end = 0,4
while start < end:
   a[start], a[end]= a[end], a[start]
   start += 1
   end -= 1
   print(a)



#15- Implement a code to find and remove duplicates from a list while preserving the original order of elements

def remove_duplicate(list1):
   return list(dict.fromkeys(list1))
list1= [1,2,3,1,1,2,4,5,6,3,4]
new_list= remove_duplicate(list1)
print(new_list)


#16- Create a code to check if a given list is sorted (either in ascending or descending order) or not.

list1=[1,2,3,4,5]
list2=[5,4,3,2,1]
#sorting list in ascending order
print(list1 == sorted(list1))
print(list2 == sorted(list2))

print(list1 == sorted(list1, reverse= "true"))
print(list2 == sorted(list2, reverse= "true"))


#17- Write a code to merge two sorted lists into a single sorted list.

a= [1,5,11,7]
b= [2,4,0,6,8]

print(a == sorted(b))
print(b == sorted(b))

single_sorted_list= sorted(a + b)
print("combine sorted list:", single_sorted_list)


#18- Implement a code to find the intersection of two given lists

def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3
lst1 = [4, 9, 1, 17, 11, 26, 28, 54, 69]
lst2 = [9, 9, 74, 21, 45, 11, 63, 28, 26]
print(intersection(lst1, lst2))


#19-  Create a code to find the union of two lists without duplicates

def union(list1, list2):
   final_list = list1 + list2
   return final_list
list1= [10,20,30,40,50,]
list2= [60,70,80,90,100]
print(union(list1, list2))

#20-  Write a code to shuffle a given list randomly without using any built-in shuffle functions.

import random
def shuffle2():
    i = 1
    lis = [5,6,7,10,11,12]
    print (lis)
    while i < len(lis):
        x = random.randint (0,len (lis)-1)
        y = random.randint (0,len (lis)-1)
        z = lis[x], lis[y] = lis[y], lis[x]
        i = i + 1
        print (lis)


#21- Write a code that takes two tuples as input and returns a new tuple containing elements that are
#     common to both input tuples

# initialize tuples
test_tup1 = (1, 3, 5)
test_tup2 = (4, 6)

# printing original tuples
print("The original tuple 1 : " + str(test_tup1))
print("The original tuple 2 : " + str(test_tup2))

# Ways to concatenate tuples
# using + operator
res = test_tup1 + test_tup2

# printing result
print("The tuple after concatenation is : " + str(res))


#22-  Create a code that prompts the user to enter two sets of integers separated by commas. Then, print the
#     intersection of these two sets

str1 = input("Enter a first set of integers seprated by commas: ")
set1 = set(map(int,str1.split(',')))
str2 = input("Enter a second set of integers seprated by commas: ")
set2 = set(map(int,str2.split(',')))
intersection = set1.intersection(set2)
print(intersection)


#23-  Write a code to concatenate two tuples. The function should take two tuples as input and return a new
#      tuple containing elements from both input tuples.

# initialize tuples
test_tup1 = (1, 3, 5)
test_tup2 = (4, 6)

# printing original tuples
print("The original tuple 1 : " + str(test_tup1))
print("The original tuple 2 : " + str(test_tup2))
res = test_tup1 + test_tup2
print("The tuple after concatenation is : " + str(res))


#24- Develop a code that prompts the user to input two sets of strings. Then, print the elements that are
#    present in the first set but not in the second set

str1 = input("Enter first string:")
str2 = input("Enter second string:")
str1 = set(str1)
str2 = set(str2)
difference = str1 - str2
print(difference)


#25- Create a code that takes a tuple and two integers as input. The function should return a new tuple
#    containing elements from the original tuple within the specified range of indices

tpl = (input("Enter a tuple: "))
tpl = tuple(tpl.split(','))
str1 = int(input("Enter interger values:"))
str2 = int(input("Enter interger values:"))
if str1 < 0 or str2 >= len(tpl) or str1>str2:
    print("Indices are out of range.")
else:
    new_tuple = tpl[str1:str2 + 1]
    print("Original Tuple:", tpl)
    print("New Tuple:", new_tuple)


#26- Write a code that prompts the user to input two sets of characters. Then, print the union of these two sets

set1 = set(input("Enter first set:"))
set2 = set(input("Enter second set:"))
unn = set1.union(set2)
print(unn)


#27-  Develop a code that takes a tuple of integers as input. The function should return the maximum and
#     minimum values from the tuple using tuple unpacking

str_lst = input("Enter integer values seprated by commas: ")
int_lst = str_lst.split(",")
tpl = tuple(map(int,int_lst))
min_tpl = min(tpl)
max_tpl = max(tpl)
print("Minimum value is :", min_tpl)
print("Maximum value is :", max_tpl)


#28- Create a code that defines two sets of integers. Then, print the union, intersection, and difference of these 
#     two sets

set1 = set([1,2,3,4,5,6])
set2 = set([4,5,6,7,8,9])
set_union = set1.union(set2)
print("Sets Union is :", set_union)
set_intersection = set1.intersection(set2)
print("Sets intersection is :", set_intersection)
set_dif = set1^set2
print("Sets difference is :", set_dif)


#29- Write a code that takes a tuple and an element as input. The function should return the count of occurrences 
#    of the given element in the tuple


def Count(tup, en):
	return tup.count(en)
tup = (10, 8, 5, 2, 10, 15, 10, 8, 5, 8, 8, 2)
enq = 4
enq1 = 10
enq2 = 8
print(Count(tup, enq), "times")
print(Count(tup, enq1), "times")
print(Count(tup, enq2), "times")

#30-  Develop a code that prompts the user to input two sets of strings. Then, print the symmetric difference of
#     these two sets

str1 = set(input(" enter a set of strings:").split())
str2 = set(input(" enter a set of strings:").split())
sym_dif = str1.symmetric_difference(str2)
print(sym_dif)


#31- Write a code that takes a list of words as input and returns a dictionary where the keys are unique words
#    and the values are the frequencies of those words in the input list


lst_tem = input("Enter the list:")
lst = set(lst_tem.split(","))
words_frqcs = {}
for word in lst:
  words = word.split()
  for i in words:
    if i in words_frqcs:
      words_frqcs[i] += 1
    else:
      words_frqcs[i] = 1
print(words_frqcs)


#32-  Write a code that takes two dictionaries as input and merges them into a single dictionary. If there are
#     common keys, the values should be added together


input1 = input("Enter the first dictionary (format: key1:value1, key2:value2): ")
input2 = input("Enter the second dictionary (format: key1:value1, key2:value2): ")

dict1 = {}
for pair in input1.split(","):
    key, value = pair.split(":")
    dict1[key.strip()] = int(value.strip())

dict2 = {}
for pair in input2.split(","):
    key, value = pair.split(":")
    dict2[key.strip()] = int(value.strip())

merged_dict = dict1.copy()

for key, value in dict2.items():
    if key in merged_dict:
        merged_dict[key] += value
    else:
        merged_dict[key] = value

print("Merged Dictionary:", merged_dict)


#33-  Write a code to access a value in a nested dictionary. The function should take the dictionary and a list of
#     keys as input, and return the corresponding value. If any of the keys do not exist in the dictionary, the
#     function should return None


nested_dict = {
    'a': {
        'b': {
            'c': 'd'
        }
    },
    'e': {
        'f': 'g'
    }
}

keys = ['a', 'b', 'c']
current_value = nested_dict

for key in keys:
    if isinstance(current_value, dict) and key in current_value:
        current_value = current_value[key]
    else:
        current_value = None
        break

print(f"The value for keys {keys} is: {current_value}")

keys = ['e', 'f']
current_value = nested_dict

for key in keys:
    if isinstance(current_value, dict) and key in current_value:
        current_value = current_value[key]
    else:
        current_value = None
        break

print(f"The value for keys {keys} is: {current_value}")

keys = ['a', 'x', 'y']
current_value = nested_dict

for key in keys:
    if isinstance(current_value, dict) and key in current_value:
        current_value = current_value[key]
    else:
        current_value = None
        break

print(f"The value for keys {keys} is: {current_value}")


#34-   Write a code that takes a dictionary as input and returns a sorted version of it based on the values. You
#       can choose whether to sort in ascending or descending order

dict_input = {'apple': 3, 'banana': 1, 'cherry': 2, 'date': 5, 'elderberry': 4}
sorted_dict_asc = dict(sorted(dict_input.items(), key=lambda item: item[1]))
print("Sorted dictionary in ascending order:", sorted_dict_asc)
sorted_dict_desc = dict(sorted(dict_input.items(), key=lambda item: item[1], reverse=True))
print("Sorted dictionary in descending order:", sorted_dict_desc)


#35-  Write a code that inverts a dictionary, swapping keys and values. Ensure that the inverted dictionary
#      correctly handles cases where multiple keys have the same value by storing the keys as a list in the
#      inverted dictionary.


original_dict = {
    'a': 1,
    'b': 2,
    'c': 1,
    'd': 3,
    'e': 2
}
inverted_dict = {}
for key, value in original_dict.items():
    if value not in inverted_dict:
        inverted_dict[value] = [key]
    else:
        inverted_dict[value].append(key)

print("Original dictionary:", original_dict)
print("Inverted dictionary:", inverted_dict)



```
